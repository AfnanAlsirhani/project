#  <span style="background-color: powderblue">Problem Statement:</span>


Coffee Corp. is a small business that aims to expand their coffee shops branches in the train stations at New York City. in this project, DSS will analyze the turnstiles data for each station to find the busiest stations. Hence, recommending the best location for the new branches so that Coffee Corp. can achieve their goals to expand their coffee business.
